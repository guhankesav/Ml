```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10.0,5.0)

data=pd.read_csv('/content/sample_data/50_Startups.csv')

#print(data)
x=data.iloc[:,:4]
#x=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
x_df = pd.DataFrame(x)
y_df = pd.DataFrame(y)
```

```python
from sklearn.preprocessing import LabelBinarizer
State_bin=LabelBinarizer()
#x=State_bin.fit_transform(x.State.values)
x_df = x_df.join(pd.DataFrame(State_bin.fit_transform(x_df.State.values),
                          columns=State_bin.classes_, 
                          index=x_df.index))
x_df1=x_df.drop(['State'], axis = 1)
x_df1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_df1,y_df,test_size=0.2)
print (x_train.shape,x_test.shape,y_train.shape,y_test.shape )
```

```python
regr = SGDRegressor(max_iter=10000)
regr.fit(x_train, y_train.values.ravel())
```

```python
y_predict=regr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test, y_predict))
```

```python
z=[[155429,345782,239828,1,0,0]];
regr.predict(z)
```

```python
print(regr.intercept_)
print(regr.coef_ )
```
