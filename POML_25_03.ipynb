```python
from sklearn.linear_model import SGDRegressor,LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler


import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10.0,5.0)

data=pd.read_csv('')

x=data.iloc[:,:4]
y=data.iloc[:,-1:]
```

```python
x_df = pd.DataFrame(x)
y_df = pd.DataFrame(y)
```

```python
from sklearn.preprocessing import LabelBinarizer
State_bin=LabelBinarizer()
x_df = x_df.join(pd.DataFrame(State_bin.fit_transform(x_df.State.values),
                          columns=State_bin.classes_, 
                          index=x_df.index))
x_df1=x_df.drop(['State'], axis = 1)
x_df1
```

```python
corr_dataset = x_df1.join(y_df)
corr_dataset
```

```python
#Using Pearson Correlation
plt.figure(figsize=(12,10))
cor = corr_dataset.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

You filter the independent variables and take only the subset of the relevant features which are linearly co-related to the dependent variable(Profit).The filtering here is done using correlation matrix and it is most commonly done using Pearson correlation.

```python
cor_target = abs(cor["Profit"])
relevant_features = cor_target[cor_target>0.5]
relevant_features
```

From the above co-relation calculation we can conclude that only R&D Spend and Marketing Spend are highly co-related with Profit so we consider only them for the training dataset

```python
x_df1=x_df1.drop(['Administration','California','Florida','New York'], axis = 1)
x_df1
```

```python
modified_dataset=x_df1.join(y_df)
modified_dataset
```

```python
print(modified_dataset[["R&D Spend","Marketing Spend"]].corr())
print(modified_dataset[["R&D Spend","Profit"]].corr())
print(modified_dataset[["Marketing Spend","Profit"]].corr())
```

 One of the assumptions of linear regression is that the independent variables need to be uncorrelated with each other. Since both R&D Spend and Marketing Spend are highly co-related with each other, we consider only one variable from those 2 for the training dataset.Hence, we consider only R&D Spend for the training dataset because they are highly co-related to Profit than Marketing Spend

```python
x_df1=x_df1.drop(['Marketing Spend'], axis = 1)
x_df1
```

In the below process, we scale the entire dataset modified after co-relation using Standard Scaler

```python
final_dataset=x_df1.join(y_df)
trans = StandardScaler()
data = pd.DataFrame(trans.fit_transform(final_dataset.values))
```

```python
x_final=data.iloc[:,:1]
y_final=data.iloc[:,-1:]
x_final
```

Independent variable(R&D Spend) after scaling is shown above

```python
y_final
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_final,y_final,test_size=0.2)
print (x_train.shape,x_test.shape,y_train.shape,y_test.shape )
```

SGDRegression model

```python
regr = SGDRegressor(max_iter=10000)
regr.fit(x_train, y_train.values.ravel())
```

```python
y_predict=regr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test, y_predict))
```

Since the R2 value is very close to 1, we can tell that this model's predictions are almost similar to the desired output.
